# Step 1 : Read the .fasta file contaning 

```python
#for read a unique string with all protein sequences concatenated-#

# NAME FUNCTION : readFastaFile1
# PARAMETERS : the filename (the file is in fasta format)
# RETURNS : a unique string with all the protein sequences concatenated

def read_FastaFile(filename) :
    
    #opening the file whose name is filename
    fd = open(filename,'r')
    txt = fd.read()
    fd.close()
    
    #.txt contains all the text of the file. 
    #seperate the proteins sequences --> the symbol that starts a new protein is '>'
    seqs = txt.split('>')[1:]
    s = ""
    
    #for each sequence in sequences
        #a lines corresponds to the line , start of first line
        #for each line in lines
            #concatenation of the followinf line with the previous one
    #Start again with the following sequence
    
    for seq in seqs :
        lines = seq.split('\n')[1:]     
        for line in lines :
            s = s + line
    return(s)

#application of the function to the file "human-proteom-small"
file_proteome = read_FastaFile("human-proteome-small.fasta")
print(file_proteome)
```

# Step 2: Read the .txt file containing the English words

```python
# NAME FUNCTION : read_file_words
# PARAMETERS : the filename (the file is in txt format)
# RETURNS : a unique string with all words

def read_file_words(filename):
    #Creation of an object which is empty
    #open a file which contain english words named f_in
        #For each line in f_in
            #Creation a line which is empty
            #add words and capitalization conversion
    #Start again with the following word
    
    list_words = [] 
    with open(filename, "r") as f_in: 
        for line in f_in: 
            word = line.strip() 
            list_words.append(word.upper())
    return list_words

#application of the function to the file "english-words.txt"
file_english_words = read_file_words("english-words.txt")
print (file_english_words)
```

# Step 3: List of words without B,J,O,U,X,Z and "-".

```python
# NAME FUNCTION : select_words_without_BJOUXZ
# PARAMETERS : the filename (the file is in txt format)
# RETURNS : a unique string with all words without the letters B, J, O, U, X and Z.

def select_words_without_BJOUXZ (filename):
    #list empty for words without BJOUXZ and "-"
    #file using is read by function read_file_words
    #vector which containt "undesirable" letter
    
    #for the word in file
        #for the condition is true
            #for each character in word
                #for each letter in letter
                    #if the character of bjouxz matches the letter of the word
                        #then the condition is false
            #if the condition is true
                #then add the words in list
        #Start again with the following word
    
    list_words_without_BJOUXZ = []
    f_in = read_file_words(filename)
    bjouxz=["B","J","O","U","X","Z","-"]
    
    for word in f_in :
        condition = True
        for character in word :
            for letter_abs in bjouxz :
                if character == letter_abs :
                    condition = False
        if condition == True :
            list_words_without_BJOUXZ.append(word)
                
    return(list_words_without_BJOUXZ)              

#application of the function to the file "english-words.txt"
file_words_without_BJOUXZ = select_words_without_BJOUXZ ("english-words.txt")
print(file_words_without_BJOUXZ)
```

# Step 4: Search words in proteome

```python
# NAME FUNCTION : search_words_in_proteome
# PARAMETERS : the file "file_english_words" and "file_proteome"
# RETURNS : two list. One list of words found in proteome and one list of number of times the word is found

def search_words_in_proteome(filename1, filename2):
    #empty list for words found
    #empty list for number of times the word is found    
    
    found_words = []
    count_found_words = []
    
    #for the word in file 1
        #if the word is in file2
            #add the word in the list found_words
            #count the words
            #add this number in the list count_found_words
    #Start again with the following word
    
    for word in filename1 :
        if word in filename2:
            found_words.append(word)
            count_word = filename2.count(word)
            count_found_words.append(count_word)
            
    return(found_words,count_found_words)

#application of the function to the files
words_in_proteome = search_words_in_proteome(file_english_words,file_proteome)
print(words_in_proteome)
```
